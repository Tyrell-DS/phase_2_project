```python
# Cleaning the Data, replacing 0 bedrooms with the mean of bedrooms, replacing 0 amd 0.5 bathrooms with mean of bathrooms. I 
# say that a home isn't a home unless there's someone to sleep, and that 0 and 0.5 bathrooms do not contain enough facilities. 
# replaced the 33 bedrooms with 3. I belive it was a user input error.
clean_dict_bathrooms = {0: int(df['bathrooms'].mean()),
                        0.5: int(df['bathrooms'].mean())}

clean_dict_rooms = {0: int(df['bedrooms'].mean())}

df['bedrooms'] = df['bedrooms'].replace(clean_dict_rooms)
df['bathrooms'] = df['bathrooms'].replace(clean_dict_bathrooms)
df['bedrooms'].replace(33,3, inplace = True)
```

```python
# Above I replaced 33 bedrooms with 3. However, I cant check through the the hold out set for any outlier. This creates upper limits
# on bedrooms and bathroom so that my data doesn't become skewed. 
df['bedrooms']= np.where(df['bedrooms'] >11, 10, df['bedrooms'])
df['bathrooms']= np.where(df['bathrooms'] >8, 7, df['bathrooms'])
```

```python
# Created Multiple new features to help explore the data set
# If a house has not been renovated Years Since last Construction is 2021 minus year built
# If a house has been renovated Years since Last Construction is 2021 - yar renovated
for label, row in df.iterrows():
    if row['yr_renovated'] ==0:
        df.loc[label,'YSLC'] = 2021 - row['yr_built']
    else: 
        df.loc[label,'YSLC'] = 2021 - row['yr_renovated'] 
#Breaks up when each house was sold on a quarterly basis        
df['sale_quarter'] = pd.DatetimeIndex(df['date']).month //4 +1  
# Divides the living space by combination of bedroomns and bathrooms to get an idea of square feet per room
df['SqFpR'] = df['sqft_living'] / (df['bedrooms'] + df['bathrooms'])
# Breaks up conditions into a categorical variable so that I can group conditions together and create dummies later on
df['new_condition'] = (df['condition']) //2 +1  
# creates a new categorical column called basement that dtermines if a house has a basement
df['bsmt'] = np.where(df['sqft_basement']>0, 1, 0)
# Tried to get an idea of yard size by diving the living space by floors and subtractiong that result from the lot size. Not used in model. 
df['yard'] = df['sqft_lot'] - (df['sqft_living']/df['floors'])
```
